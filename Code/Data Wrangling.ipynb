## Loading our Independent Variables

```python
# Loading Data into a DF named vdem
vdem = pd.read_csv(r"Data\V-Dem-CY-Core-v11.1.csv")

# Printing the first five rows of the data set
vdem.head(250)
```

### Dropping Columns



```python
vdem = vdem.filter(['country_name', 'country_text_id', 'year', 'v2x_polyarchy', 'v2x_freexp_altinf', 'v2xel_frefair', 'v2xcl_rol', 'v2x_jucon', 'v2xeg_eqprotec'])
```

```python
vdem.head(10)
```

```python
vdem_29 = pd.DataFrame({'year': [1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]})
```

```python
vdem_29 = vdem[vdem['year'].isin([1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018])]
```

```python
vdem_29.head(30)
```

```python
vdem_29 = vdem_29.rename(columns={'country_text_id': 'cnt'})
```

```python
vdem_29.head(5)
```

```python
etd_15.head(5)
```

### Merging the Dataframes

```python
merged_df = vdem_29.merge(etd_15, how = 'inner', on = ['cnt', 'year'])
```

```python
merged_df.head(50)
```

```python
merged_df.shape
```

### Check the Data

```python
# Group by state
merged_countries = merged_df.groupby('cnt')

# Check for latest date
merged_countries.last()
```

```python
merged_countries.last().shape
```
