### Merging the Dataframes

### Check the Data

```python
# Group by state
merged_countries = merged_df.groupby('cnt')

# Check for latest date
merged_countries.last()
```

```python
merged_countries.last().shape
```

```python
merged_df[['FinSec Growth']]=merged_df.sort_values(['year']).groupby('cnt')[['Financial services']].pct_change()
merged_df
```
